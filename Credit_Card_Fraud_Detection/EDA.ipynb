# Exploratory Data Analysis (EDA)

```python
# Import necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the cleaned data
df = pd.read_csv(r"C:\Summer Training\Projects\Final Project\Data\cleaned_creditcard.csv")

# Preview the dataset
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Visualize class distribution
sns.countplot(x='Class', data=df)
plt.title('Class Distribution (0 = Non-Fraud, 1 = Fraud)')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()

# Percentage of each class
class_percent = df['Class'].value_counts(normalize=True) * 100
print(class_percent)
```

```python
# Correlation matrix
corr_matrix = df.corr()

plt.figure(figsize=(20,15))
sns.heatmap(corr_matrix, cmap='coolwarm_r', annot=True)
plt.title('Correlation Matrix of All Features')
plt.show()
```

```python
# Correlation of features with target class
corr_with_class = corr_matrix['Class'].sort_values(ascending=False)
print("Top features positively correlated with fraud:")
print(corr_with_class.head(10))

print("\nTop features negatively correlated with fraud:")
print(corr_with_class.tail(10))
```

```python
# Distribution of a few important features
features_to_plot = ['V14', 'V10', 'V17', 'V12', 'scaled_amount'] # highest coorelation columns

for feature in features_to_plot:
    plt.figure(figsize=(8,4))
    sns.kdeplot(df[feature][df['Class'] == 0], label='Non-Fraud')
    sns.kdeplot(df[feature][df['Class'] == 1], label='Fraud')
    plt.title(f'Distribution of {feature} by Class')
    plt.legend()
    plt.show()
```

```python
# Skewness and kurtosis for numerical features
from scipy.stats import skew, kurtosis

skew_vals = df.drop('Class', axis=1).apply(skew)
kurt_vals = df.drop('Class', axis=1).apply(kurtosis)

skew_df = pd.DataFrame({'Feature': skew_vals.index, 'Skewness': skew_vals.values, 'Kurtosis': kurt_vals.values})
print(skew_df.sort_values(by='Skewness', key=np.abs, ascending=False).head(10))
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
components = pca.fit_transform(df.drop('Class', axis=1))
plt.figure(figsize=(8,6))
sns.scatterplot(x=components[:,0], y=components[:,1], hue=df['Class'], alpha=0.7)
plt.title('2D PCA Visualization of Fraud vs Non-Fraud')
plt.show()
```
